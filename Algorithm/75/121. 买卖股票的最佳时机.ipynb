### LeetCode 题目：121. 买卖股票的最佳时机

https://leetcode.com/problems/best-time-to-buy-and-sell-stock/description/

#### 问题描述：
给定一个数组 `prices`，其中 `prices[i]` 表示某个股票在第 `i` 天的价格。你想在某一天买入这只股票，并在未来的某一天卖出。计算你可以获得的最大利润。如果你不能获得利润，返回 0。

#### 示例 1：
```python
输入：prices = [7, 1, 5, 3, 6, 4]
输出：5
```

#### 示例 2：
```python
输入：prices = [7, 6, 4, 3, 1]
输出：0
```

### 原始代码：

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0  # 初始化最大利润为 0
        buy = prices[0]  # 设置初始买入价格为第一天的价格
        for p in prices[1:]:  # 遍历价格列表，从第二天开始
            if p > buy:  # 如果当前价格高于买入价格
                max_profit = max(max_profit, p - buy)  # 更新最大利润
            else:  # 如果当前价格低于或等于买入价格
                buy = p  # 更新买入价格为当前价格
        return max_profit  # 返回最大利润
```

### 改进后的代码：

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:  # 如果价格列表为空，返回 0
            return 0

        max_profit = 0  # 初始化最大利润为 0
        buy = float('inf')  # 将买入价格初始化为正无穷大

        for p in prices:  # 遍历所有价格
            if p < buy:  # 更新买入价格为更低的价格
                buy = p
            elif p - buy > max_profit:  # 检查当前价格减去买入价格是否产生更高利润
                max_profit = p - buy  # 更新最大利润
        
        return max_profit  # 返回最大利润
```

### 改进点解释：

1. **初始化买入价格**：
   - 在改进后的代码中，买入价格 `buy` 初始化为正无穷大，这样可以处理价格列表为空的情况，并确保第一个价格会成为新的买入价格。

2. **遍历逻辑简化**：
   - 在改进后的代码中，遍历价格列表时不再从第二天开始，而是从第一天开始，这样更加简洁和易读。

3. **利润计算**：
   - 在比较当前价格与买入价格时，直接更新最大利润的逻辑更加清晰。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码遍历了 `prices` 列表一次，因此时间复杂度是 O(n)，其中 n 是 `prices` 列表的长度。

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间来存储 `max_profit` 和 `buy`，因此空间复杂度是 O(1)。

### 总结：

- **贪心算法**：通过始终更新最低买入价格和当前最大利润，保持对最佳收益的追踪。
- **简洁性与效率**：改进后的代码在逻辑上更简洁，且不影响性能，反而提高了可读性。
