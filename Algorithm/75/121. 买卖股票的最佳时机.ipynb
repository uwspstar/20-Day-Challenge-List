#### 问题描述：
给定一个数组 `prices`，其中 `prices[i]` 表示第 `i` 天的股票价格。你想最大化你的利润，只能进行一次买卖操作（即先买入然后卖出）。返回你能获取的最大利润。如果无法获取利润，返回 0。

#### 示例 1：
```python
输入：prices = [7, 1, 5, 3, 6, 4]
输出：5
解释：买入价格为 1，卖出价格为 6，利润为 6 - 1 = 5。
```

#### 示例 2：
```python
输入：prices = [7, 6, 4, 3, 1]
输出：0
解释：在这种情况下，没有交易完成，利润为 0。
```

### 原始代码（一次遍历法）：

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        buy = prices[0]  # 初始化买入价格为第一天的价格
        max_pft = 0  # 初始化最大利润为 0
        
        for p in prices[1:]:  # 从第二天开始遍历价格
            if p > buy:  # 如果当前价格大于买入价格
                max_pft = max(max_pft, p - buy)  # 计算可能的最大利润
            else:
                buy = p  # 更新买入价格为当前价格
        return max_pft  # 返回最大利润
```

### 逐行解释与示例运行：

1. **`def maxProfit(self, prices: List[int]) -> int:`**
   - 定义主函数 `maxProfit`，接受一个整数列表 `prices`，返回最大利润。

2. **`buy = prices[0]`**
   - **初始化买入价格为第一天的价格**。

3. **`max_pft = 0`**
   - **初始化最大利润为 0**。

4. **`for p in prices[1:]:`**
   - 遍历从第二天开始的价格列表。

5. **`if p > buy:`**
   - **如果当前价格大于买入价格**。

6. **`max_pft = max(max_pft, p - buy)`**
   - **更新最大利润**，计算当前价格减去买入价格的利润。

7. **`else:`**
   - 如果当前价格不高于买入价格。

8. **`buy = p`**
   - **更新买入价格为当前价格**。

9. **`return max_pft`**
   - **返回最大利润**。

   - `p = 6`：`6 > 1`，更新 `max_pft = max(4, 6 - 1)` → `max_pft = 5`。
   - `p = 4`：`4 > 1`，保持 `max_pft = 5`。

3. **返回结果**：
   - 返回 `max_pft = 5`。

### 改进方案（使用动态规划）：

虽然当前实现已经很高效，但可以通过动态规划的方式来理解问题。

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:  # 检查输入是否为空
            return 0
        
        min_price = float('inf')  # 初始化最小价格为无穷大
        max_profit = 0  # 初始化最大利润为 0

        for price in prices:  # 遍历价格
            if price < min_price:  # 更新最小价格
                min_price = price
            elif price - min_price > max_profit:  # 更新最大利润
                max_profit = price - min_price
        
        return max_profit  # 返回最大利润
```

### 改进后的代码：

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:  # 如果价格列表为空，返回 0
            return 0

        max_profit = 0  # 初始化最大利润为 0
        buy = float('inf')  # 将买入价格初始化为正无穷大

        for p in prices:  # 遍历所有价格
            if p < buy:  # 更新买入价格为更低的价格
                buy = p
            elif p - buy > max_profit:  # 检查当前价格减去买入价格是否产生更高利润
                max_profit = p - buy  # 更新最大利润
        
        return max_profit  # 返回最大利润
```

### 改进点解释：

1. **初始化买入价格**：
   - 在改进后的代码中，买入价格 `buy` 初始化为正无穷大，这样可以处理价格列表为空的情况，并确保第一个价格会成为新的买入价格。

2. **遍历逻辑简化**：
   - 在改进后的代码中，遍历价格列表时不再从第二天开始，而是从第一天开始，这样更加简洁和易读。

3. **利润计算**：
   - 在比较当前价格与买入价格时，直接更新最大利润的逻辑更加清晰。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码遍历了 `prices` 列表一次，因此时间复杂度是 O(n)，其中 n 是 `prices` 列表的长度。

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间来存储 `max_profit` 和 `buy`，因此空间复杂度是 O(1)。
