### LeetCode 题目：11. 盛最多水的容器

#### 原始代码（迭代方法）：

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_area = 0  # 初始化最大面积为 0
        l, r = 0, len(height) - 1  # 左指针 l 从数组开头开始，右指针 r 从数组末尾开始
        
        while l < r:  # 当左指针小于右指针时，继续计算面积
            area = (r - l) * min(height[l], height[r])  # 计算当前容器的面积
            max_area = max(area, max_area)  # 更新最大面积，如果当前面积更大
            
            if height[l] < height[r]:  # 比较左右两边的高度
                l += 1  # 移动左指针，寻找可能的更高的容器
            else:
                r -= 1  # 移动右指针，寻找可能的更高的容器
        
        return max_area  # 返回找到的最大面积
```

### 逐行解释：

1. **`max_area = 0`**：
   - 初始化变量 `max_area` 为 0，用于存储找到的最大水容量。

2. **`l, r = 0, len(height) - 1`**：
   - `l` 为左指针，初始化为数组的起始索引 0。
   - `r` 为右指针，初始化为数组的最后一个索引。

3. **`while l < r:`**：
   - 进入循环，当左指针小于右指针时进行面积计算。这个条件确保了有容器的存在。

4. **`area = (r - l) * min(height[l], height[r])`**：
   - 计算当前容器的水容量。容器的宽度为 `r - l`，高度为两根柱子中较短的那根，使用 `min(height[l], height[r])`。

5. **`max_area = max(area, max_area)`**：
   - 更新 `max_area`，取当前计算的面积和之前最大面积的较大值。

6. **`if height[l] < height[r]:`**：
   - 判断左边的柱子是否矮于右边的柱子。如果是，则移动左指针。

7. **`l += 1`**：
   - 移动左指针到右侧一位，准备下一次计算。移动较短的柱子可能找到更大的容器。

8. **`else:`**：
   - 如果右边的柱子矮于或等于左边的柱子，则移动右指针。

9. **`r -= 1`**：
   - 移动右指针到左侧一位，准备下一次计算。

10. **`return max_area`**：
    - 循环结束后，返回找到的最大水容量。

### 递归方法代码：

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        def calculate_area(l: int, r: int) -> int:
            if l >= r:  # 基本情况，左指针与右指针重合
                return 0  # 如果重合，返回 0
            
            area = (r - l) * min(height[l], height[r])  # 计算当前容器的面积
            
            # 递归选择移动较短的柱子
            if height[l] < height[r]:
                return max(area, calculate_area(l + 1, r))  # 移动左指针并计算剩余面积
            else:
                return max(area, calculate_area(l, r - 1))  # 移动右指针并计算剩余面积
        
        return calculate_area(0, len(height) - 1)  # 开始递归
```

2. **`if l >= r:`**：
   - 基本情况：如果左指针大于或等于右指针，表示没有容器可以计算。

3. **`return 0`**：

4. **`area = (r - l) * min(height[l], height[r])`**：

5. **`if height[l] < height[r]:`**：
   - 判断左边的柱子是否矮于右边的柱子。

6. **`return max(area, calculate_area(l + 1, r))`**：
   - 如果左边的柱子矮，递归移动左指针并计算剩余的水容量，返回当前面积和递归计算的较大值。

7. **`else:`**：
   - 如果右边的柱子矮于或等于左边的柱子。

8. **`return max(area, calculate_area(l, r - 1))`**：
   - 递归移动右指针并计算剩余的水容量，返回当前面积和递归计算的较大值。

9. **`return calculate_area(0, len(height) - 1)`**：
