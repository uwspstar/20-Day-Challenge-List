### LeetCode 题目：572. 另一棵树的子树

https://leetcode.com/problems/subtree-of-another-tree/description/

https://www.youtube.com/watch?v=Ja7TdUIW-_Q&list=PLx2fZyxEuihLSjtEzLKxf20yV8LKLAibL&index=21

#### 问题描述：
给定两棵二叉树 `root` 和 `subRoot`，检查 `subRoot` 是否是 `root` 的子树。一个树 `t` 是另一棵树 `s` 的子树，当 `s` 中存在一棵树，与 `t` 的结构和节点值完全相同。

#### 示例 1：
```python
输入：root = [3,4,5,1,2], subRoot = [4,1,2]
输出：true
```

#### 示例 2：
```python
输入：root = [3,4,5,1,2,null,null,null,null,0], subRoot = [4,1,2]
输出：false
```

### 原始代码（递归方法）：

```python
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        if not root:  # 如果 root 为空，subRoot 不可能是它的子树
            return False

        if self.is_same_tree(root, subRoot):  # 检查当前节点是否相同
            return True
        
        # 递归检查左子树和右子树
        return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)

    
    def is_same_tree(self, p, q):
        if not p and not q:  # 如果两个树都为空，返回 True
            return True
        if not p or not q:  # 如果只有一个树为空，返回 False
            return False
        if p.val != q.val:  # 如果节点值不同，返回 False
            return False
        # 递归检查左右子树
        return self.is_same_tree(p.left, q.left) and self.is_same_tree(p.right, q.right)
```

### 逐行解释：

#### `isSubtree` 方法

1. **`def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:`**：
   - 定义主函数 `isSubtree`，接受两个二叉树的根节点 `root` 和 `subRoot`，返回布尔值表示 `subRoot` 是否是 `root` 的子树。

2. **`if not root:`**：
   - 检查 `root` 是否为空。如果是，返回 `False`，因为空树不能包含任何子树。

3. **`if self.is_same_tree(root, subRoot):`**：
   - 调用 `is_same_tree` 方法检查当前节点 `root` 和 `subRoot` 是否相同。如果相同，返回 `True`。

4. **`return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)`**：
   - 递归检查 `root` 的左子树和右子树，看 `subRoot` 是否是它们的子树。如果任一子树中存在 `subRoot`，返回 `True`。

#### `is_same_tree` 方法

1. **`def is_same_tree(self, p, q):`**：
   - 定义辅助函数 `is_same_tree`，用于检查两棵树是否相同。

2. **`if not p and not q:`**：
   - 检查两个树是否都为空。如果是，返回 `True`。

3. **`if not p or not q:`**：
   - 检查是否只有一个树为空。如果是，返回 `False`。

4. **`if p.val != q.val:`**：
   - 检查当前节点的值是否相同。如果不同，返回 `False`。

5. **`return self.is_same_tree(p.left, q.left) and self.is_same_tree(p.right, q.right)`**：
   - 递归检查左右子树，只有在左右子树都相同的情况下返回 `True`。

### 时间复杂度分析：

- **时间复杂度**：O(m * n)
  - 其中 m 是 `root` 的节点数，n 是 `subRoot` 的节点数。最坏情况下，我们需要遍历整个 `root` 树，并对每个节点调用 `is_same_tree`。

- **空间复杂度**：O(h)
  - 递归调用栈的深度最大为树的高度 h，因此空间复杂度为 O(h)。

### 总结：

- **递归方法**：通过递归比较两棵树的结构和节点值，逻辑清晰，易于理解。
- **高效性**：虽然在时间复杂度上不是最优，但该方法在处理一般情况时表现良好。
