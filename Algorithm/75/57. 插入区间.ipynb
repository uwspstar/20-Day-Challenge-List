### LeetCode 题目：57. 插入区间

https://leetcode.com/problems/insert-interval/description/

https://www.youtube.com/watch?v=_1jHnaovdpo&list=PLx2fZyxEuihLSjtEzLKxf20yV8LKLAibL&index=20

#### 问题描述：
给定一个不重叠的区间列表 `intervals` 和一个新插入的区间 `newInterval`，请将 `newInterval` 合并到 `intervals` 中，并返回合并后的区间列表（保证结果中的区间仍然是排序的，不重叠的）。

#### 示例 1：
```python
输入：intervals = [[1,3],[6,9]], newInterval = [2,5]
输出：[[1,5],[6,9]]
```

#### 示例 2：
```python
输入：intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]], newInterval = [4,8]
输出：[[1,2],[3,10],[12,16]]
```

### 原始代码（插入并合并）：

```python
class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        res = []  # 用于存储合并后的区间
        i = 0
        
        # 添加所有在 newInterval 之前的区间
        while i < len(intervals) and intervals[i][1] < newInterval[0]:
            res.append(intervals[i])
            i += 1
        
        # 合并所有与 newInterval 重叠的区间
        while i < len(intervals) and intervals[i][0] <= newInterval[1]:
            newInterval[0] = min(newInterval[0], intervals[i][0])  # 更新新区间的开始
            newInterval[1] = max(newInterval[1], intervals[i][1])  # 更新新区间的结束
            i += 1
        
        res.append(newInterval)  # 添加合并后的新区间

        # 添加所有在 newInterval 之后的区间
        while i < len(intervals):
            res.append(intervals[i])
            i += 1
        
        return res  # 返回合并后的区间
```

### 逐行解释：

1. **`def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:`**：
   - 定义主函数 `insert`，接受不重叠的区间列表 `intervals` 和一个新插入的区间 `newInterval`，返回合并后的区间列表。

2. **`res = []`**：
   - 初始化一个空列表 `res`，用于存储合并后的区间。

3. **`i = 0`**：
   - 初始化索引 `i`，用于遍历 `intervals`。

4. **`while i < len(intervals) and intervals[i][1] < newInterval[0]:`**：
   - 这个循环用于将所有在 `newInterval` 之前的区间添加到 `res` 中。
   - 当当前区间的结束时间小于 `newInterval` 的开始时间时，表示没有重叠，直接添加。

5. **`res.append(intervals[i])`**：
   - 将当前区间添加到结果列表 `res`。

6. **`i += 1`**：
   - 移动到下一个区间。

7. **`while i < len(intervals) and intervals[i][0] <= newInterval[1]:`**：
   - 这个循环用于合并所有与 `newInterval` 重叠的区间。
   - 当当前区间的开始时间小于等于 `newInterval` 的结束时间时，表示有重叠。

8. **`newInterval[0] = min(newInterval[0], intervals[i][0])`**：
   - 更新 `newInterval` 的开始时间为与当前区间的开始时间的较小值。

9. **`newInterval[1] = max(newInterval[1], intervals[i][1])`**：
   - 更新 `newInterval` 的结束时间为与当前区间的结束时间的较大值。

10. **`i += 1`**：
    - 移动到下一个区间。

11. **`res.append(newInterval)`**：
    - 将合并后的 `newInterval` 添加到结果列表 `res`。

12. **`while i < len(intervals):`**：
    - 这个循环用于将所有在 `newInterval` 之后的区间添加到 `res` 中。

13. **`res.append(intervals[i])`**：
    - 将当前区间添加到结果列表 `res`。

14. **`i += 1`**：
    - 移动到下一个区间。

15. **`return res`**：
    - 返回合并后的区间列表。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 在最坏情况下，所有区间都需要被遍历，因此时间复杂度为 O(n)，其中 n 是区间的数量。

- **空间复杂度**：O(n)
  - 在最坏情况下，所有区间都不重叠，因此需要存储所有区间在结果列表中。

### 总结：

- **排序与合并**：通过先遍历并添加所有不重叠的区间，再合并重叠的区间，最后将合并后的区间添加到结果中。
- **高效性**：该算法能够有效处理较大的输入，并快速找到并合并重叠的区间。
