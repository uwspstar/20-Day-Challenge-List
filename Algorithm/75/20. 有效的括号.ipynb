### LeetCode 题目：20. 有效的括号

https://leetcode.com/problems/valid-parentheses/description/

#### 问题描述：
给定一个只包括 `'('`，`')'`，`'{'`，`'}'`，`'['`，`']'` 的字符串，判断字符串是否有效。有效字符串需满足：
- 左括号必须用相同类型的右括号闭合。
- 左括号必须以正确的顺序闭合。

#### 示例 1：
```python
输入：s = "()"
输出：true
```

#### 示例 2：
```python
输入：s = "(]"
输出：false
```

### 原始代码：

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stk = []  # 创建一个栈用于存储左括号
        pairs = {"(": ")", "[": "]", "{": "}"}  # 定义括号对
        
        for c in s:  # 遍历字符串中的每个字符
            if c in pairs:  # 如果字符是左括号
                stk.append(c)  # 将其推入栈中
            else:  # 如果字符是右括号
                # 检查栈是否为空或当前右括号与栈顶左括号不匹配
                if len(stk) == 0 or c != pairs[stk.pop()]:
                    return False  # 返回 False，表示无效
        return not stk  # 返回 True，如果栈为空，表示有效
```

### 逐行解释：

1. **`stk = []`**：
   - 初始化一个空栈 `stk` 用于存储左括号，以便后续进行匹配。

2. **`pairs = {"(": ")", "[": "]", "{": "}"}`**：
   - 定义一个字典 `pairs`，用于存储左括号与其对应的右括号的映射关系。

3. **`for c in s:`**：
   - 遍历字符串 `s` 中的每个字符 `c`。

4. **`if c in pairs:`**：
   - 检查当前字符 `c` 是否是左括号。

5. **`stk.append(c)`**：
   - 如果是左括号，将其推入栈 `stk`。

6. **`else:`**：
   - 如果字符 `c` 不是左括号，即为右括号。

7. **`if len(stk) == 0 or c != pairs[stk.pop()]:`**：
   - 检查栈是否为空，或者当前右括号与栈顶的左括号不匹配。
     - `len(stk) == 0`：表示没有左括号可以匹配。
     - `c != pairs[stk.pop()]`：表示当前右括号与栈顶左括号不匹配。

8. **`return False`**：
   - 如果上面的条件成立，返回 `False`，表示字符串无效。

9. **`return not stk`**：
   - 遍历完成后，检查栈是否为空。栈为空表示所有左括号都有匹配的右括号，返回 `True`；否则返回 `False`。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码遍历字符串一次，因此时间复杂度为 O(n)，其中 n 是字符串 `s` 的长度。

- **空间复杂度**：O(n)
  - 在最坏情况下，栈中可能存储所有左括号，因此空间复杂度为 O(n)。

### 总结：

- **栈的使用**：通过使用栈结构，可以有效地管理括号的匹配关系，确保每个左括号都有相应的右括号。
- **高效性**：该算法在时间和空间上的效率使其非常适合处理括号匹配问题。
