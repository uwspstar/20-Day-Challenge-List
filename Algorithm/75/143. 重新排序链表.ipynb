### LeetCode 问题：143. 重新排序链表

https://leetcode.com/problems/reorder-list/description/
https://www.youtube.com/watch?v=S5bfdUTrKLM&t=2s

#### 问题描述：
给定一个链表的头节点，您需要按以下方式重新排序它：
- 您必须原地执行此操作，而不改变节点的值。
- 重新排序的链表应遵循以下模式：
  - `L0 → Ln → L1 → Ln-1 → L2 → Ln-2 → ...`

例如，给定链表：
- `1 → 2 → 3 → 4 → 5`
重新排序后的链表应为：
- `1 → 5 → 2 → 4 → 3`

#### 示例 1：
```python
输入：head = [1, 2, 3, 4]
输出：[1, 4, 2, 3]
```

#### 示例 2：
```python
输入：head = [1, 2, 3, 4, 5]
输出：[1, 5, 2, 4, 3]
```

### 原始代码（迭代方法）：

```python
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        if not head:
            return head  # 如果链表为空，直接返回
        
        slow, fast = head, head.next  # 使用快慢指针找到链表的中间节点

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        cur = slow.next  # 从中间节点的下一个节点开始
        pre = slow.next = None  # 将后半部分的头节点分离，初始化前一个指针为 None

        while cur:  # 反转链表的后半部分
            nxt = cur.next  # 存储下一个节点
            cur.next = pre  # 反转当前节点的链接
            pre = cur  # 更新前一个节点
            cur = nxt  # 移动到下一个节点
        
        l1, l2 = head, pre  # l1 为前半部分，l2 为反转后的后半部分

        while l2:  # 交替合并两个链表
            tp1, tp2 = l1.next, l2.next  # 记录 l1 和 l2 的下一个节点
            l1.next = l2  # 将 l2 节点连接到 l1
            l2.next = tp1  # 将 l1 的下一个节点连接到 tp1
            l1, l2 = tp1, tp2  # 移动到下一个节点
```

### 行逐行解释与示例执行：

1. **`def reorderList(self, head: Optional[ListNode]) -> None:`**
   - 定义主函数 `reorderList`，接受链表的头节点并原地重新排序。

2. **`if not head:`**
   - 检查链表是否为空。如果是，直接返回。

3. **`slow, fast = head, head.next`**
   - 使用快慢指针初始化，`slow` 指向链表头，`fast` 指向下一个节点。

4. **`while fast and fast.next:`**
   - 使用快慢指针找到链表的中间节点，`slow` 指针在最后停留的位置就是中间节点。

5. **`cur = slow.next`**
   - 将 `cur` 指向中间节点的下一个节点，作为后半部分的开始。

6. **`pre = slow.next = None`**
   - 分离后半部分，设置中间节点的下一个节点为 `None`，初始化前一个指针为 `None`。

7. **`while cur:`**
   - 反转链表的后半部分，迭代直到 `cur` 为 `None`。

8. **`nxt = cur.next`**
   - 存储当前节点的下一个节点，以便在反转链接后使用。

9. **`cur.next = pre`**
   - 反转当前节点的链接，使其指向前一个节点。

10. **`pre = cur`**
    - 将前一个指针移动到当前节点，为下一次迭代做准备。

11. **`cur = nxt`**
    - 移动当前指针到下一个节点。

12. **`l1, l2 = head, pre`**
    - 初始化 `l1` 指向前半部分的头节点，`l2` 指向反转后的后半部分的头节点。

13. **`while l2:`**
    - 合并两个链表，直到 `l2` 为 `None`。

14. **`tp1, tp2 = l1.next, l2.next`**
    - 在合并之前，记录 `l1` 和 `l2` 的下一个节点。

15. **`l1.next = l2`**
    - 将当前的 `l2` 节点连接到 `l1`。

16. **`l2.next = tp1`**
    - 将 `l1` 的下一个节点连接到 `tp1`。

17. **`l1, l2 = tp1, tp2`**
    - 移动 `l1` 和 `l2` 指针到它们的下一个节点，继续合并。

### 示例执行：

假设我们有以下链表：

```python
head = 1 -> 2 -> 3 -> 4 -> 5 -> None
```

执行步骤：

1. **初始调用**：
   - `head = 1`
   - `slow = 1`，`fast = 2`

2. **找到中间节点**：
   - **1st iteration**: `slow = 2`，`fast = 4`
   - **2nd iteration**: `slow = 3`，`fast = None`
   - 此时 `slow` 指向 `3`，`cur` 初始化为 `4`。

3. **反转后半部分**：
   - **1st iteration**: `cur = 4`，`nxt = 5`，`4.next = None` → 链表变为 `4 -> None`
   - **2nd iteration**: `cur = 5`，`nxt = None`，`5.next = 4` → 链表变为 `5 -> 4`
   - 后半部分现在被反转为 `5 -> 4`。

4. **合并两个部分**：
   - **1st iteration**: `l1 = 1`，`l2 = 5`，连接 `1 -> 5`，再连接 `5 -> 2`。
   - **2nd iteration**: `l1 = 2`，`l2 = 4`，连接 `2 -> 4`，再连接 `4 -> 3`。
   - **3rd iteration**: `l1 = 3`，`l2 = None`，完成合并。

5. **最终结果**：
   - 重新排序后的链表为 `1 -> 5 -> 2 -> 4 -> 3 -> None`。
