### LeetCode 题目：15. 三数之和

https://leetcode.com/problems/3sum/description/

#### 问题描述：
给定一个整数数组 `nums`，判断是否存在三个元素 `a, b, c` 使得 `a + b + c = 0`。找出所有满足条件且不重复的三元组。

#### 示例 1：
```python
输入：nums = [-1,0,1,2,-1,-4]
输出：[[-1,-1,2],[-1,0,1]]
```

#### 示例 2：
```python
输入：nums = []
输出：[]
```

### 原始代码（双指针法）：

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()  # 对数组进行排序
        ans = []  # 初始化结果列表
        
        for i in range(len(nums) - 2):  # 遍历数组，至倒数第三个元素
            if i > 0 and nums[i] == nums[i - 1]:  # 跳过重复的元素
                continue
            
            l = i + 1  # 左指针初始化
            r = len(nums) - 1  # 右指针初始化
            while l < r:  # 当左指针小于右指针时
                total = nums[i] + nums[l] + nums[r]  # 计算三数之和
                if total > 0:  # 如果和大于0，移动右指针向左
                    r -= 1
                elif total < 0:  # 如果和小于0，移动左指针向右
                    l += 1
                else:  # 如果和等于0，找到一个三元组
                    tmp = [nums[i], nums[l], nums[r]]  # 存储找到的三元组
                    ans.append(tmp)  # 添加结果
                    while l < r and nums[l] == tmp[1]:  # 跳过重复的左指针元素
                        l += 1
                    while l < r and nums[r] == tmp[2]:  # 跳过重复的右指针元素
                        r -= 1
        
        return ans  # 返回结果
```

### 逐行解释：

1. **`def threeSum(self, nums: List[int]) -> List[List[int]]:`**：
   - 定义主函数 `threeSum`，接受整数数组 `nums`，返回一个包含三元组的列表。

2. **`nums.sort()`**：
   - 对数组进行排序，以便使用双指针法查找三元组。

3. **`ans = []`**：
   - 初始化结果列表 `ans`，用于存储满足条件的三元组。

4. **`for i in range(len(nums) - 2):`**：
   - 遍历数组，直到倒数第三个元素。因为要找三个数，所以需要留出两个位置。

5. **`if i > 0 and nums[i] == nums[i - 1]:`**：
   - 跳过重复的元素，以避免添加重复的三元组。

6. **`l = i + 1`**：
   - 初始化左指针 `l` 为当前元素的下一个元素。

7. **`r = len(nums) - 1`**：
   - 初始化右指针 `r` 为数组最后一个元素。

8. **`while l < r:`**：
   - 当左指针小于右指针时，进入循环。

9. **`total = nums[i] + nums[l] + nums[r]`**：
   - 计算当前三数之和。

10. **`if total > 0:`**：
    - 如果和大于0，说明需要更小的数，移动右指针向左。

11. **`r -= 1`**：
    - 移动右指针到左侧一位。

12. **`elif total < 0:`**：
    - 如果和小于0，说明需要更大的数，移动左指针向右。

13. **`l += 1`**：
    - 移动左指针到右侧一位。

14. **`else:`**：
    - 如果和等于0，找到了一个三元组。

15. **`tmp = [nums[i], nums[l], nums[r]]`**：
    - 存储找到的三元组。

16. **`ans.append(tmp)`**：
    - 将找到的三元组添加到结果列表中。

17. **`while l < r and nums[l] == tmp[1]:`**：
    - 跳过重复的左指针元素，避免添加重复的三元组。

18. **`while l < r and nums[r] == tmp[2]:`**：
    - 跳过重复的右指针元素，避免添加重复的三元组。

19. **`return ans`**：
    - 返回满足条件的所有三元组。

### 时间复杂度分析：

- **时间复杂度**：O(n^2)
  - 外层循环 O(n)，内层双指针遍历 O(n)，整体复杂度为 O(n^2)，其中 n 是数组 `nums` 的长度。

- **空间复杂度**：O(1)
  - 不使用额外的数组，结果存储在 `ans` 中，不考虑输入的大小，因此空间复杂度为 O(1)。

### 总结：

- **双指针法**：通过排序和双指针有效查找三元组，避免了使用三重循环的低效。
- **高效性**：该算法在处理较大输入时表现良好，并且能够避免重复的三元组。

在这段代码中，使用条件 `if i > 0 and nums[i] == nums[i - 1]:` 的目的是为了跳过重复的元素，以避免在结果中出现重复的三元组。以下是一个例子来解释为什么不使用 `if i > 0 and nums[i] == nums[i + 1]:`。

### 示例：

假设我们有一个排序后的数组：
```python
nums = [-1, -1, 0, 1, 2, 2]
```

#### 1. 使用 `if i > 0 and nums[i] == nums[i - 1]:`

当我们遍历这个数组时，`i` 的值可能会是 1。在这个位置，`nums[i]` 是 `-1`，而 `nums[i - 1]` 也是 `-1`，因此条件为真。这意味着我们已经处理了这个 `-1` 作为第一个元素的情况。如果不跳过这个重复的 `-1`，我们会再次尝试找到以这个重复的 `-1` 开头的三元组，这会导致我们重复记录同样的三元组，比如 `[-1, -1, 2]`。

#### 2. 不使用 `if i > 0 and nums[i] == nums[i + 1]:`

如果我们使用 `if i > 0 and nums[i] == nums[i + 1]:`，假设 `i` 的值为 1，此时 `nums[i]` 是 `-1`，而 `nums[i + 1]` 是 `0`。在这种情况下，条件不成立，导致我们不会跳过 `-1`，从而可能在后续的循环中再次考虑相同的元素作为三元组的第一个元素。

### 结论

- 使用 `if i > 0 and nums[i] == nums[i - 1]:` 是为了确保每个数字只在需要时被考虑一次，避免重复。
- 使用 `if i > 0 and nums[i] == nums[i + 1]:` 将无法有效地跳过已处理的数字，可能导致重复的三元组被添加到结果中。

因此，选择检查前一个元素是否相等是合理的，而不是检查下一个元素。这样可以确保生成的结果中不包含重复的三元组。
