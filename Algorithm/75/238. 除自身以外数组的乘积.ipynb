### LeetCode 题目：238. 除自身以外数组的乘积

https://leetcode.com/problems/product-of-array-except-self/description/

#### 问题描述：
给定一个整数数组 `nums`，返回一个数组 `answer`，使得 `answer[i]` 等于 `nums` 中除 `nums[i]` 以外的所有元素的乘积。

#### 示例 1：
```python
输入：nums = [1,2,3,4]
输出：[24,12,8,6]
```

#### 示例 2：
```python
输入：nums = [-1,1,0,-3,3]
输出：[0,0,9,0,0]
```

### 原始代码：

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]: 
        n = len(nums)  # 获取数组的长度
        ans = [1] * n  # 初始化结果数组，所有元素初始为 1
        
        # 第一个循环，计算左侧乘积
        for i in range(1, n):
            ans[i] = ans[i - 1] * nums[i - 1]  # ans[i] 为 nums[i-1] 之前所有数字的乘积
        
        tmp = nums[-1]  # 将最后一个元素赋值给 tmp，用于右侧乘积计算
        # 第二个循环，计算右侧乘积
        for i in range(n - 2, -1, -1):
            ans[i] *= tmp  # 将当前索引的结果与右侧的乘积相乘
            tmp *= nums[i]  # 更新 tmp 为当前元素的乘积，以便下一次循环使用
        
        return ans  # 返回最终结果数组
```

### 逐行解释：

1. **`def productExceptSelf(self, nums: List[int]) -> List[int]:`**：
   - 定义主函数 `productExceptSelf`，接受整数数组 `nums`，返回一个整数数组。

2. **`n = len(nums)`**：
   - 获取数组 `nums` 的长度，并存储在变量 `n` 中。

3. **`ans = [1] * n`**：
   - 初始化结果数组 `ans`，其长度与 `nums` 相同，所有元素初始为 1。

4. **`for i in range(1, n):`**：
   - 遍历数组 `nums` 从索引 1 到 `n-1`，计算左侧乘积。

5. **`ans[i] = ans[i - 1] * nums[i - 1]`**：
   - 对于每个索引 `i`，将 `ans[i]` 设为 `ans[i - 1]`（到 `i-1` 的所有元素的乘积）乘以 `nums[i - 1]`（`i` 左侧所有元素的乘积）。

6. **`tmp = nums[-1]`**：
   - 将最后一个元素 `nums[-1]` 赋值给变量 `tmp`，用于后续右侧乘积的计算。

7. **`for i in range(n - 2, -1, -1):`**：
   - 从倒数第二个元素开始，向前遍历数组（从 `n-2` 到 `0`），计算右侧乘积。

8. **`ans[i] *= tmp`**：
   - 将当前索引的 `ans[i]` 与 `tmp` 相乘，更新 `ans[i]` 为当前元素乘以右侧元素的乘积。

9. **`tmp *= nums[i]`**：
   - 更新 `tmp`，将当前元素 `nums[i]` 加入到 `tmp` 中，为下一次循环准备右侧乘积。

10. **`return ans`**：
    - 返回最终的结果数组 `ans`。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码中有两个循环，每个循环遍历数组一次，因此时间复杂度为 O(n)，其中 n 是数组 `nums` 的长度。

- **空间复杂度**：O(1)
  - 使用了常量级的额外空间来存储 `tmp`，不考虑结果数组 `ans`，因此空间复杂度为 O(1)。

### 总结：

- **左侧和右侧乘积计算**：通过两次遍历，分别计算左侧和右侧的乘积，避免了使用除法，解决了数组中元素为 0 的情况。
- **高效性**：该算法在时间和空间上的效率使其非常适合处理大规模输入。
