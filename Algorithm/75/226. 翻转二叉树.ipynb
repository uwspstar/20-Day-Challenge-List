#### 问题描述：
翻转一棵二叉树。

#### 示例：
```python
输入：
     4
   /   \
  2     7
 / \   / \
1   3 6   9

输出：
     4
   /   \
  7     2
 / \   / \
9   6 3   1
```

### 原始代码（递归方法）：

```python
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:  # 如果树为空，直接返回
            return root
        
        lft = self.invertTree(root.left)  # 递归翻转左子树
        rgt = self.invertTree(root.right)  # 递归翻转右子树

        root.left, root.right = rgt, lft  # 交换左右子树

        return root  # 返回翻转后的树根
```

### 逐行解释与关键点：

1. **`def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:`**：
   - 定义主函数 `invertTree`，接受一个二叉树的根节点 `root`，返回翻转后的树根。

2. **`if not root:`**：
   - **关键点**：检查树是否为空。如果是，直接返回 `root`，以处理递归的基本情况。

3. **`lft = self.invertTree(root.left)`**：
   - 递归调用翻转左子树，并将结果保存到 `lft` 中。

4. **`rgt = self.invertTree(root.right)`**：
   - 递归调用翻转右子树，并将结果保存到 `rgt` 中。

5. **`root.left, root.right = rgt, lft`**：
   - **关键点**：交换当前节点的左子树和右子树，实现翻转。

6. **`return root`**：
   - 返回翻转后的树根。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - **关键点**：每个节点被访问一次，因此时间复杂度为 O(n)，其中 n 是树中的节点数。

- **空间复杂度**：O(h)
  - **关键点**：递归调用栈的深度最大为树的高度 h，因此空间复杂度为 O(h)。

### 总结：

- **递归方法**：通过递归有效地翻转二叉树，逻辑清晰且易于实现。
- **高效性**：该算法能够处理较大的树结构，快速完成翻转操作。

### 应用技巧

1. **递归思维**：
   - 在处理树形结构时，递归是一个非常自然的选择，尤其是在需要处理左右子树的场景中。

2. **基础情况处理**：
   - 在递归中，总是要考虑基础情况，以避免无限递归。

3. **左右子树的交换**：
   - 在许多树形操作中，左右子树的交换是常见操作，理解其背后的逻辑非常重要。
