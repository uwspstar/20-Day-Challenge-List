### LeetCode 题目：56. 合并区间

https://leetcode.com/problems/merge-intervals/description/

https://www.youtube.com/watch?v=fm6bYjYQX_A&list=PLx2fZyxEuihLSjtEzLKxf20yV8LKLAibL&index=19

#### 问题描述：
给定一个区间的集合，请合并所有重叠的区间。

#### 示例 1：
```python
输入：intervals = [[1,3],[2,6],[8,10],[15,18]]
输出：[[1,6],[8,10],[15,18]]
```

#### 示例 2：
```python
输入：intervals = [[1,4],[4,5]]
输出：[[1,5]]
```

### 原始代码（排序和合并）：

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        intervals.sort(key=lambda x: x[0])  # 按照区间的起始时间排序
        merged = []  # 用于存储合并后的区间

        for interval in intervals:  # 遍历每个区间
            # 如果 merged 为空或没有重叠，直接添加当前区间
            if not merged or merged[-1][1] < interval[0]:
                merged.append(interval)
            else:
                # 存在重叠，合并当前区间与最后一个区间
                merged[-1][1] = max(merged[-1][1], interval[1])

        return merged  # 返回合并后的区间
```

### 逐行解释：

1. **`def merge(self, intervals: List[List[int]]) -> List[List[int]]:`**：
   - 定义主函数 `merge`，接受一个区间的列表 `intervals`，返回合并后的区间列表。

2. **`intervals.sort(key=lambda x: x[0])`**：
   - 按照每个区间的起始时间进行排序，以便于后续合并操作。

3. **`merged = []`**：
   - 初始化一个空列表 `merged`，用于存储合并后的区间。

4. **`for interval in intervals:`**：
   - 遍历每个排序后的区间。

5. **`if not merged or merged[-1][1] < interval[0]:`**：
   - 如果 `merged` 为空，或者当前区间的起始时间大于上一个合并区间的结束时间，表示没有重叠，直接将当前区间添加到 `merged`。

6. **`merged.append(interval)`**：
   - 添加当前区间到 `merged`。

7. **`else:`**：
   - 如果存在重叠，执行合并操作。

8. **`merged[-1][1] = max(merged[-1][1], interval[1])`**：
   - 更新上一个合并区间的结束时间，取当前区间和上一个合并区间结束时间的较大值。

9. **`return merged`**：
   - 返回合并后的区间列表。

### 示例过程

假设我们有以下输入：

```python
intervals = [[1, 3], [2, 6], [8, 10], [15, 18]]
```

1. **排序后的区间**：
   ```
   [[1, 3], [2, 6], [8, 10], [15, 18]]
   ```

2. **合并过程**：
   - 遍历第一个区间 `[1, 3]`，因为 `merged` 为空，添加到 `merged`：
     ```
     merged = [[1, 3]]
     ```
   - 遍历第二个区间 `[2, 6]`，与 `[1, 3]` 有重叠，合并为 `[1, 6]`：
     ```
     merged = [[1, 6]]
     ```
   - 遍历第三个区间 `[8, 10]`，没有重叠，直接添加：
     ```
     merged = [[1, 6], [8, 10]]
     ```
   - 遍历第四个区间 `[15, 18]`，没有重叠，直接添加：
     ```
     merged = [[1, 6], [8, 10], [15, 18]]
     ```

### 时间复杂度分析：

- **时间复杂度**：O(n log n)
  - 排序需要 O(n log n) 的时间，合并过程遍历区间需要 O(n) 的时间，因此总时间复杂度为 O(n log n)。

- **空间复杂度**：O(n)
  - 在最坏情况下，所有区间都不重叠，需要存储所有区间，因此空间复杂度为 O(n)。

### 总结：

- **排序与合并**：通过先对区间进行排序，再逐个合并相重叠的区间，逻辑清晰且高效。
- **高效性**：该算法能够有效处理较大的输入，并且能够快速找到并合并重叠的区间。
