### 1. LeetCode 1: 两数之和 (Two Sum)

#### Problem Description  
给定一个整数数组 `nums` 和一个目标值 `target`，请在该数组中找出和为目标值的两个整数，并返回它们的数组下标。可以假设每种输入只会对应一个答案，且不能重复使用相同的元素。

#### 解法：哈希表  
我们可以通过哈希表在一次遍历中找到两个数，使得它们的和为目标值 `target`。每次遍历时，检查当前数的补数是否已经存在于哈希表中。

#### Python 代码：

```python
def twoSum(nums, target):
    hash_map = {}  # 初始化哈希表
    for i, num in enumerate(nums):
        complement = target - num  # 计算补数
        if complement in hash_map:  # 检查补数是否存在于哈希表中
            return [hash_map[complement], i]  # 如果存在，返回补数和当前数的下标
        hash_map[num] = i  # 将当前数存入哈希表
```

##### 解释：
- `hash_map` 是一个哈希表，用于存储已遍历过的数的值和下标。
- 每次遍历数组时，计算当前数的补数 `complement`，并检查该补数是否已经存在于哈希表中。
- 如果找到补数，则返回它们的下标；如果未找到，则将当前数存入哈希表中供后续使用。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(n)

### 2. LeetCode 121: 买卖股票的最佳时机 (Best Time to Buy and Sell Stock)

#### Problem Description  
给定一个数组 `prices`，其中 `prices[i]` 是某支股票第 `i` 天的价格。你只能选择一天买入这只股票，并选择未来的某一天卖出，计算你所能获取的最大利润。如果你不能获取任何利润，返回 0。

#### 解法：一次遍历  
我们只需要遍历价格数组，并记录到目前为止的最低价格，并计算当前价格与最低价格的差值，来更新最大利润。

#### Python 代码：

```python
def maxProfit(prices):
    min_price = float('inf')  # 初始化最低价格为无穷大
    max_profit = 0  # 初始化最大利润为 0
    for price in prices:
        if price < min_price:
            min_price = price  # 更新最低价格
        elif price - min_price > max_profit:
            max_profit = price - min_price  # 更新最大利润
    return max_profit
```

##### 解释：
- 通过一次遍历，记录当前的最低价格 `min_price`，同时计算当前价格与最低价格的差值，并更新最大利润 `max_profit`。
- 如果当前价格低于最低价格，则更新最低价格。
- 如果当前利润大于之前的最大利润，则更新最大利润。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(1)

### 3. LeetCode 217: 存在重复元素 (Contains Duplicate)

#### Problem Description  
给定一个整数数组 `nums`，判断是否存在重复元素。如果任意一个值在数组中出现至少两次，返回 `true`；如果数组中每个元素都不相同，返回 `false`。

#### 解法：哈希表  
我们可以使用哈希表来检查数组中是否有重复元素。遍历数组，每次检查当前元素是否已经存在于哈希表中，如果存在，则返回 `true`；否则将其加入哈希表。

#### Python 代码：

```python
def containsDuplicate(nums):
    hash_set = set()  # 使用集合来存储元素
    for num in nums:
        if num in hash_set:  # 如果当前元素已经存在于集合中
            return True
        hash_set.add(num)  # 否则将当前元素加入集合
    return False
```

##### 解释：
- 使用集合 `hash_set` 来存储已经遍历过的元素。
- 如果当前元素已经在集合中存在，则表示数组中存在重复元素，返回 `true`。
- 如果遍历完数组没有发现重复元素，则返回 `false`。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(n)

### 4. LeetCode 238: 除自身以外数组的乘积 (Product of Array Except Self)

#### Problem Description  
给定一个整数数组 `nums`，返回一个数组 `answer`，其中 `answer[i]` 是 `nums` 中除了 `nums[i]` 之外其他元素的乘积。请在不使用除法的情况下完成此题。

#### 解法：前缀与后缀乘积  
我们可以通过前缀和后缀的乘积来计算结果。首先遍历数组计算每个元素的前缀乘积，然后再反向遍历计算后缀乘积并更新结果。

#### Python 代码：

```python
def productExceptSelf(nums):
    n = len(nums)
    result = [1] * n  # 初始化结果数组
    prefix = 1  # 前缀乘积
    for i in range(n):
        result[i] = prefix  # 记录前缀乘积
        prefix *= nums[i]  # 更新前缀乘积
    suffix = 1  # 后缀乘积
    for i in range(n - 1, -1, -1):
        result[i] *= suffix  # 更新结果数组为前缀乘积 * 后缀乘积
        suffix *= nums[i]  # 更新后缀乘积
    return result
```

##### 解释：
- 第一次遍历计算前缀乘积，并存储在结果数组中。
- 第二次反向遍历计算后缀乘积，并将其与前缀乘积相乘更新结果。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(1)（不包括结果数组）

### 5. LeetCode 53: 最大子数组和 (Maximum Subarray)

#### Problem Description  
给定一个整数数组 `nums`，找到具有最大和的连续子数组，并返回其和。

#### 解法：动态规划  
我们可以通过动态规划的思想来解决这个问题。对于每个元素，决定是将它加入前面的子数组，还是从它开始一个新的子数组。

#### Python 代码：

```python
def maxSubArray(nums):
    current_sum = max_sum = nums[0]  # 初始化当前子数组和和最大子数组和
    for num in nums[1:]:
        current_sum = max(num, current_sum + num)  # 更新当前子数组和
        max_sum = max(max_sum, current_sum)  # 更新最大子数组和
    return max_sum
```

##### 解释：
- `current_sum` 表示当前子数组的和，每次决定是否将当前元素加入当前子数组，还是从当前元素开始新的子数组。
- `max_sum` 记录迄今为止的最大子数组和。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(1)

### 6. LeetCode 152: 乘积最大子数组 (Maximum Product Subarray)

#### Problem Description  
给定一个整数数组 `nums`，找出其中乘积最大的连续子数组，并返回该乘积。

#### 解法：动态规划  
我们需要同时记录当前子数组的最大乘积和最小乘积，因为负数乘负数可能会变成最大正数。

#### Python 代码：

```python
def maxProduct(nums):
    max_prod = min_prod = result = nums[0]  # 初始化最大乘积、最小乘积和结果
    for num in nums[1:]:
        if num < 0:
            max_prod, min_prod = min_prod, max_prod  # 如果当前数为负数，交换最大和最小乘积
        max_prod = max(num, max_prod * num)  # 更新最大乘积
        min_prod = min(num, min_prod * num)  # 更新最小乘积
        result = max(result, max_prod)  # 更新结果
    return result
```

##### 解释：
- 对于每个元素，计算它可能的最大乘积和最小乘积。
- 如果当前元素为负数，最大和最小乘积交换，因为负数乘以最小乘积可能会变成最大乘积。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(1)

### 7. LeetCode 153: 寻找旋转排序数组中的最小值 (Find Minimum in Rotated Sorted Array)

#### Problem Description  
给定一个旋转排序数组，找到其中的最小值。假设数组中不存在重复元素。

#### 解法：二分查找  
使用二分查找法，通过比较中间元素与右侧元素，来缩小查找范围，直到找到最小值。

#### Python 代码：

```python
def findMin(nums):
    left, right = 0, len(nums) - 1  # 初始化左右指针
    while left < right:
        mid = (left + right) // 2  # 计算中间索引
        if nums[mid] > nums[right]:  # 如果中间元素大于右侧元素，说明最小值在右侧
            left = mid + 1
        else:
            right = mid  # 否则最小值在左侧
    return nums[left]
```

##### 解释：
- 通过二分查找的方式，不断缩小搜索区间，直到找到数组的最小值。
- 每次通过比较中间元素与右侧元素来判断最小值所在的区间。

##### 时间复杂度：O(log n)  
##### 空间复杂度：O(1)

### 8. LeetCode 33: 搜索旋转排序数组 (Search in Rotated Sorted Array)

#### Problem Description  
给定一个旋转排序数组和一个目标值，在数组中搜索目标值，如果存在返回其索引，否则返回 -1。假设数组中不存在重复元素。

#### 解法：二分查找  
我们可以通过修改二分查找的方法，在旋转排序数组中找到目标值。

#### Python 代码：

```python
def search(nums, target):
    left, right = 0, len(nums) - 1  # 初始化左右指针
    while left <= right:
        mid = (left + right) // 2  # 计算中间索引
        if nums[mid] == target:  # 如果找到目标值
            return mid
        if nums[left]

 <= nums[mid]:  # 左半部分有序
            if nums[left] <= target < nums[mid]:
                right = mid - 1  # 目标值在左半部分
            else:
                left = mid + 1  # 否则在右半部分
        else:  # 右半部分有序
            if nums[mid] < target <= nums[right]:
                left = mid + 1  # 目标值在右半部分
            else:
                right = mid - 1  # 否则在左半部分
    return -1
```

##### 解释：
- 通过修改二分查找的方法，在每次比较中判断当前区间的有序部分。
- 根据有序部分的范围，判断目标值是否在该区间内，从而缩小查找范围。

##### 时间复杂度：O(log n)  
##### 空间复杂度：O(1)

### 9. LeetCode 15: 三数之和 (3Sum)

#### Problem Description  
给定一个包含 n 个整数的数组 `nums`，判断 `nums` 中是否存在三个元素使得它们的和为 0。请找出所有和为 0 且不重复的三元组。

#### 解法：双指针  
我们可以首先对数组进行排序，然后使用双指针的方法来查找所有和为 0 的三元组。

#### Python 代码：

```python
def threeSum(nums):
    nums.sort()  # 先对数组进行排序
    result = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:  # 跳过重复元素
            continue
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                result.append([nums[i], nums[left], nums[right]])  # 找到一组解
                while left < right and nums[left] == nums[left + 1]:
                    left += 1  # 跳过重复元素
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1  # 跳过重复元素
                left += 1
                right -= 1
    return result
```

##### 解释：
- 通过对数组进行排序，然后使用双指针的方式，查找所有符合条件的三元组。
- 跳过重复元素，避免结果中出现重复的三元组。

##### 时间复杂度：O(n^2)  
##### 空间复杂度：O(n)

### 10. LeetCode 11: 盛最多水的容器 (Container With Most Water)

#### Problem Description  
给定一个数组 `height`，其中 `height[i]` 表示第 `i` 个柱子的高度。找出两根柱子组成的容器可以容纳的最多的水。

#### 解法：双指针  
我们可以使用双指针的方法，每次将较短的柱子往内移动，计算当前的容积，并更新最大容积。

#### Python 代码：

```python
def maxArea(height):
    left, right = 0, len(height) - 1  # 初始化左右指针
    max_area = 0  # 初始化最大容积
    while left < right:
        width = right - left  # 计算当前宽度
        max_area = max(max_area, min(height[left], height[right]) * width)  # 更新最大容积
        if height[left] < height[right]:
            left += 1  # 移动较短的柱子
        else:
            right -= 1
    return max_area
```

##### 解释：
- 使用双指针从数组的两端开始，逐步向内移动较短的柱子，更新可以容纳的最大水量。
- 每次计算当前容器的容积，并记录最大值。

##### 时间复杂度：O(n)  
##### 空间复杂度：O(1)
