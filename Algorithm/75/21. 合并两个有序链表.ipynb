### LeetCode 题目：21. 合并两个有序链表

https://leetcode.com/problems/merge-two-sorted-lists/description/

#### 问题描述：
将两个升序链表合并为一个新的 升序 链表，返回合并后的链表。新链表是通过拼接给定的两个链表的所有节点组成的。

#### 示例 1：
```python
输入：list1 = 1->2->4, list2 = 1->3->4
输出：1->1->2->3->4->4
```

#### 示例 2：
```python
输入：list1 = [], list2 = []
输出：[]
```

### 迭代方法的原始代码：

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = tail = ListNode()  # 创建一个虚拟节点作为合并链表的头
        l1, l2 = list1, list2
        
        while l1 and l2:  # 当两个链表都不为空时
            if l1.val < l2.val:  # 比较两个链表当前节点的值
                tail.next = l1  # 将较小的节点链接到新链表
                l1 = l1.next  # 移动到下一个节点
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next  # 更新尾节点
        
        tail.next = l1 or l2  # 将剩余节点链接到新链表
        return dummy.next  # 返回合并后的链表，跳过虚拟头节点
```

### 递归方法的代码：

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:  # 如果 list1 为空，直接返回 list2
            return list2
        if not list2:  # 如果 list2 为空，直接返回 list1
            return list1
        
        # 选择较小的节点作为当前合并链表的头
        if list1.val < list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)  # 递归合并剩余部分
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)  # 递归合并剩余部分
            return list2
```

### 递归方法解释：

1. **基本情况**：
   - 如果其中一个链表为空，则返回另一个链表（已经是合并的结果）。

2. **递归逻辑**：
   - 比较两个链表的当前节点值，选择较小的节点作为新的合并链表的头，并递归调用合并其余部分。

### 时间复杂度分析：

- **时间复杂度**：
  - 迭代和递归方法均为 O(n + m)，其中 n 和 m 是两个链表的长度。

- **空间复杂度**：
  - 迭代方法的空间复杂度为 O(1)。
  - 递归方法的空间复杂度为 O(n) 或 O(m)，因为递归调用栈的深度取决于较短链表的长度。

### 总结：

- **迭代与递归**：两种方法各有优缺点，迭代方法更高效地使用空间，而递归方法更易于理解和实现。
- **链表合并技巧**：无论哪种方法，使用比较操作来合并两个有序链表，保持了整体的高效性。
