### LeetCode 题目：206. 反转链表

https://leetcode.com/problems/reverse-linked-list/description/

https://www.youtube.com/watch?v=G0_I-ZF0S38

#### 问题描述：
反转一个单链表。

#### 示例 1：
```python
输入：head = [1,2,3,4,5]
输出：[5,4,3,2,1]
```

#### 示例 2：
```python
输入：head = [1,2]
输出：[2,1]
```

### 原始代码（递归方法）：

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head  # 如果链表为空或只有一个节点，直接返回当前节点

        new_head = self.reverseList(head.next)  # 递归调用，反转后续链表
        head.next.next = head  # 将当前节点指向前一个节点
        head.next = None  # 将当前节点的下一个节点设为 None

        return new_head  # 返回新的头节点
```

### 逐行解释：

1. **`def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode:`**：
   - 定义主函数 `reverseList`，接受单链表的头节点 `head`，返回反转后的链表头节点。

2. **`if not head or not head.next:`**：
   - 检查当前节点是否为空或者只有一个节点。如果是，说明没有节点可以反转，直接返回当前节点。

3. **`new_head = self.reverseList(head.next)`**：
   - 递归调用 `reverseList`，将链表的其余部分反转，并将结果保存到 `new_head`。这时 `new_head` 是反转后链表的头节点。

4. **`head.next.next = head`**：
   - 将当前节点的下一个节点（`head.next`）的 `next` 指向当前节点，从而将当前节点反转过来。

5. **`head.next = None`**：
   - 将当前节点的 `next` 指向 `None`，确保反转后的链表正确结束。

6. **`return new_head`**：
   - 返回反转后链表的新头节点。

### 递归思路解析

- **递归的基本思想**：通过递归深入到链表的尾部，然后在返回的过程中逐步反转指针。
- **反转过程**：
  - 先反转后续的链表部分，最后将当前节点连接到反转后的链表上。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 递归需要遍历链表的每个节点，因此时间复杂度为 O(n)，其中 n 是链表中的节点数。

- **空间复杂度**：O(n)
  - 由于递归调用栈的深度最大为 n，因此空间复杂度为 O(n)。

### 总结：

- **递归方法**：通过递归有效地反转单链表，逻辑简单且易于理解。
- **高效性**：在处理较大链表时表现良好，尽管在空间使用上会有一定消耗。

`head.next.next = head` 这一行代码的作用是将当前节点（`head`）指向它的下一个节点的 `next` 属性，从而实现链表的反转。我们来通过具体的例子进行说明。

### 示例说明

假设我们有以下链表：

```
1 -> 2 -> 3 -> None
```

在这个链表中，`1` 是头节点，`2` 是 `1` 的下一个节点，`3` 是 `2` 的下一个节点。我们要反转这个链表，使其变为：

```
3 -> 2 -> 1 -> None
```

#### 过程解析

1. **初始状态**：
   - `head` 当前指向节点 `1`，即 `head.val = 1`。
   - `head.next` 指向节点 `2`，即 `head.next.val = 2`。
   - `head.next.next` 指向节点 `3`，即 `head.next.next.val = 3`。

2. **递归调用**：
   - 我们会先调用 `reverseList(head.next)`，也就是传入 `2`。在这个递归中，`head` 现在指向 `2`。
   - 再调用 `reverseList(head.next)`，传入 `3`。在这个递归中，`head` 现在指向 `3`。
   - 当我们到达 `None` 时，递归返回，开始反转链表。

3. **执行 `head.next.next = head`**：
   - 当处理节点 `3` 时，`head` 当前指向 `3`，它的 `next` 是 `None`：
     - `head.next` 为空，所以不做任何操作。
   - 当返回到节点 `2` 时，`head` 当前指向 `2`：
     - 此时执行 `head.next.next = head`，即 `2.next.next = 2`，这将 `3` 的 `next` 指向 `2`：
       ```
       3 -> 2
       ```

4. **继续反转**：
   - 然后返回到节点 `1`，此时 `head` 当前指向 `1`：
     - 执行 `head.next.next = head`，即 `2.next.next = 1`，这将 `2` 的 `next` 指向 `1`：
       ```
       2 -> 1
       ```

5. **最终结果**：
   - 最终链表变为：
   ```
   3 -> 2 -> 1 -> None
   ```

### 总结

- **作用**：`head.next.next = head` 将当前节点指向其下一个节点的 `next` 属性，从而完成反转的连接。
- **逻辑**：通过递归的方式，逐步将每个节点的指向反转，最终实现链表的整体反转。

### 示例 2：更长的链表

```
1 -> 2 -> 3 -> 4 -> 5 -> None
```

#### 反转过程

1. **初始状态**：
   - `head` 当前指向 `1`。

2. **递归调用**：
   - 递归依次调用 `reverseList(2)`, `reverseList(3)`, `reverseList(4)`, `reverseList(5)`，最终到达 `None`。

   - 在处理节点 `5` 时，什么都不做。

     - `3.next.next = 3`，现在 `4 -> 3`。
   - 回到节点 `2` 时：
     - `2.next.next = 2`，现在 `3 -> 2`。
   - 回到节点 `1` 时：
     - `1.next.next = 1`，现在 `2 -> 1`。

   5 -> 4 -> 3 -> 2 -> 1 -> None

### 示例 3：包含一个节点的链表

#### 反转过程

1. **初始状态**：
   - `head` 当前指向 `1`。

2. **递归调用**：
   - 调用 `reverseList(None)`，返回 `head`。

3. **无反转**：
   - 因为只有一个节点，所以 `head.next` 为 `None`，不需要反转。

4. **最终结果**：
   - 链表仍然是：
   ```
   1 -> None
   ```

### 总结

- **反转逻辑**：`head.next.next = head` 在链表反转过程中至关重要。它将当前节点指向前一个节点，完成反转。
- **递归过程**：通过递归深入到链表的末尾，逐层反转每个节点的指向。
- **适用情况**：无论链表的长度是 1、2、还是更多，反转的逻辑都是相同的。
