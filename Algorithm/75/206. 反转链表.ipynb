#### 问题描述：
给定一个单链表的头节点，反转链表并返回反转后的链表。

#### 示例 1：
```python
输入：head = [1, 2, 3, 4, 5]
输出：[5, 4, 3, 2, 1]
```

#### 示例 2：
```python
输入：head = [1, 2]
输出：[2, 1]
```

#### 示例 3：
```python
输入：head = []
输出：[]
```

### 带注释的原始代码：

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:  # 基本情况：如果链表为空或只有一个节点
            return head
        
        pre, cur = None, head  # 初始化前一个指针为 None，当前指针为头节点
        # [0, 1, 2, 3]
        while cur:  # 遍历链表
            nxt = cur.next  # 存储下一个节点 1
            cur.next = pre  # 反转链接 None
            pre = cur  # 将前一个指针移动到当前节点 0
            cur = nxt  # 将当前指针移动到下一个节点 1

        return pre  # 返回反转后的新头节点
```

### 行逐行解释与示例执行：

1. **`def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:`**
   - 定义主函数 `reverseList`，接受链表的头节点并返回反转后的链表头节点。

2. **`if not head or not head.next:`**
   - 检查链表是否为空或只有一个节点。如果是，直接返回 `head`，因为它已经是反转的状态。

3. **`pre, cur = None, head`**
   - 初始化两个指针：`pre`（前一个指针）为 `None`，`cur`（当前指针）为 `head`。

4. **`while cur:`**
   - 当 `cur` 不为 `None` 时，遍历链表。

5. **`nxt = cur.next`**
   - 将下一个节点存储在 `nxt` 中。

6. **`cur.next = pre`**
   - 反转链接，使当前节点指向前一个节点。

7. **`pre = cur`**
   - 将 `pre` 指针移动到当前节点，为下一次迭代做准备。

8. **`cur = nxt`**
   - 将 `cur` 指针移动到下一个节点。

9. **`return pre`**
   - 循环结束后，`pre` 将是反转链表的新头节点，因此返回 `pre`。

### 示例执行：

   - **第 1 次迭代**：

     - `nxt = 3`

   - **第 5 次迭代**：
     - `nxt = None`
     - `5.next = 4`
     - `pre = 5`，`cur = None`

3. **最终结果**：
   - 反转后的链表为 `5 -> 4 -> 3 -> 2 -> 1 -> None`。

### 递归方法（可选）

以下是递归方法的实现以作比较：

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:  # 基本情况：如果链表为空或只有一个节点
            return head

        newhead = self.reverseList(head.next)  # 递归调用以反转其余链表
        head.next.next = head  # 使下一个节点指向当前节点
        head.next = None  # 将当前节点的下一个指针设为 None

        return newhead  # 返回反转后的新头节点
```

`head.next.next = head` 这一行代码的作用是将当前节点（`head`）指向它的下一个节点的 `next` 属性，从而实现链表的反转。我们来通过具体的例子进行说明。

### 示例说明

假设我们有以下链表：

```
1 -> 2 -> 3 -> None
```

在这个链表中，`1` 是头节点，`2` 是 `1` 的下一个节点，`3` 是 `2` 的下一个节点。我们要反转这个链表，使其变为：

```
3 -> 2 -> 1 -> None
```

#### 过程解析

1. **初始状态**：
   - `head` 当前指向节点 `1`，即 `head.val = 1`。
   - `head.next` 指向节点 `2`，即 `head.next.val = 2`。
   - `head.next.next` 指向节点 `3`，即 `head.next.next.val = 3`。

2. **递归调用**：
   - 我们会先调用 `reverseList(head.next)`，也就是传入 `2`。在这个递归中，`head` 现在指向 `2`。
   - 再调用 `reverseList(head.next)`，传入 `3`。在这个递归中，`head` 现在指向 `3`。
   - 当我们到达 `None` 时，递归返回，开始反转链表。

3. **执行 `head.next.next = head`**：
   - 当处理节点 `3` 时，`head` 当前指向 `3`，它的 `next` 是 `None`：
     - `head.next` 为空，所以不做任何操作。
   - 当返回到节点 `2` 时，`head` 当前指向 `2`：
     - 此时执行 `head.next.next = head`，即 `2.next.next = 2`，这将 `3` 的 `next` 指向 `2`：
       ```
       3 -> 2
       ```

4. **继续反转**：
   - 然后返回到节点 `1`，此时 `head` 当前指向 `1`：
     - 执行 `head.next.next = head`，即 `2.next.next = 1`，这将 `2` 的 `next` 指向 `1`：
       ```
       2 -> 1
       ```

5. **最终结果**：
   - 最终链表变为：
   ```
   3 -> 2 -> 1 -> None
   ```

### 总结

- **作用**：`head.next.next = head` 将当前节点指向其下一个节点的 `next` 属性，从而完成反转的连接。
- **逻辑**：通过递归的方式，逐步将每个节点的指向反转，最终实现链表的整体反转。
