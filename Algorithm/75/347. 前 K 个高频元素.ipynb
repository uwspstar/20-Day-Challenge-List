### LeetCode 题目：347. 前 K 个高频元素

https://leetcode.com/problems/top-k-frequent-elements/description/

https://www.youtube.com/watch?v=YPTqKIgVk-k

#### 问题描述：
给定一个整数数组 `nums` 和一个整数 `k`，返回出现频率前 `k` 个元素的列表。

#### 示例 1：
```python
输入：nums = [1,1,1,2,2,3], k = 2
输出：[1,2]
```

#### 示例 2：
```python
输入：nums = [1], k = 1
输出：[1]
```

### 原始代码（桶排序法）：

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = {}  # 创建一个字典用于计数
        freq = [[] for i in range(len(nums) + 1)]  # 创建一个桶列表，索引表示频率

        for n in nums:  # 遍历数组，计数每个数字的频率
            count[n] = count.get(n, 0) + 1
        
        for n, c in count.items():  # 将数字放入对应频率的桶中
            freq[c].append(n)

        res = []  # 初始化结果列表
        for i in range(len(freq) - 1, 0, -1):  # 从高频到低频遍历桶
            for n in freq[i]:  # 遍历当前频率的数字
                res.append(n)  # 添加到结果列表
                if len(res) == k:  # 如果结果列表已达到 k 个元素
                    return res  # 返回结果
```

### 逐行解释与示例运行：

1. **`def topKFrequent(self, nums: List[int], k: int) -> List[int]:`**
   - 定义主函数 `topKFrequent`，接受一个整数列表 `nums` 和一个整数 `k`，返回前 `k` 个高频元素。

2. **`count = {}`**
   - **创建一个字典 `count`**，用于记录每个数字出现的频率。

3. **`freq = [[] for i in range(len(nums) + 1)]`**
   - **创建一个桶列表 `freq`**，其索引表示频率，每个索引处存储频率对应的数字。

4. **`for n in nums:`**
   - 遍历数组 `nums`。

5. **`count[n] = count.get(n, 0) + 1`**
   - **更新数字的频率**，如果数字不存在于 `count` 中，默认为 0。

6. **`for n, c in count.items():`**
   - 遍历字典 `count` 中的每个数字及其频率。

7. **`freq[c].append(n)`**
   - **将数字添加到其频率对应的桶中**。

8. **`res = []`**
   - **初始化结果列表 `res`**，用于存储前 `k` 个高频元素。

9. **`for i in range(len(freq) - 1, 0, -1):`**
   - 从最高频率开始遍历桶。

10. **`for n in freq[i]:`**
    - 遍历当前频率的所有数字。

11. **`res.append(n)`**
    - **将数字添加到结果列表**。

12. **`if len(res) == k:`**
    - **检查结果列表是否已达到 `k` 个元素**。

13. **`return res`**
    - **返回结果列表**。

### 示例运行：

假设我们有以下输入：

```python
nums = [1, 1, 1, 2, 2, 3]
k = 2
```

运行代码步骤如下：

1. **初始化**：
   - `count = {}`, `freq = [[], [], [], [], [], []]`（长度为 7）。

2. **计数频率**：
   - 遍历 `nums`：
     - `1` → `count = {1: 1}`
     - `1` → `count = {1: 2}`
     - `1` → `count = {1: 3}`
     - `2` → `count = {1: 3, 2: 1}`
     - `2` → `count = {1: 3, 2: 2}`
     - `3` → `count = {1: 3, 2: 2, 3: 1}`

3. **填充频率桶**：
   - 遍历 `count`：
     - `1` → `freq[3].append(1)` → `freq = [[], [], [], [1], [], []]`
     - `2` → `freq[2].append(2)` → `freq = [[], [], [2], [1], [], []]`
     - `3` → `freq[1].append(3)` → `freq = [[3], [], [2], [1], [], []]`

4. **提取结果**：
   - `res = []`。
   - 从高频到低频遍历：
     - `i = 6`，无元素。
     - `i = 5`，无元素。
     - `i = 4`，无元素。
     - `i = 3`，`res = [1]`。
     - `i = 2`，`res = [1, 2]`（达到 `k`，返回 `res`）。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 其中 n 是数组的长度。遍历数组和桶的操作都是线性的。

### 空间复杂度分析：

- **空间复杂度**：O(n)
  - 存储频率计数和结果列表的额外空间。

### 提示和警告：

1. **边界情况**：
   - 处理输入为空或 `k` 为 0 的情况，返回空列表。

2. **数字范围**：
   - 由于使用了桶排序法，确保数字的范围合理，以免浪费空间。

3. **大数据处理**：
   - 对于非常大的输入，考虑使用其他数据结构（如最小堆）来优化空间使用。

### 改进方案（使用最小堆）：

可以使用最小堆来维护前 k 个高频元素，更适合处理频率不均匀的情况。

```python
from collections import Counter
import heapq

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = Counter(nums)  # 计数每个数字的频率
        return heapq.nlargest(k, count.keys(), key=count.get)  # 返回频率前 k 的数字
```

### 示例运行（改进方案）：

假设我们有以下输入：

```python
nums = [1, 1, 1, 2, 2, 3]
k = 2
```

运行代码步骤如下：

1. **计数频率**：
   - `count = Counter(nums)` 结果为 `Counter({1: 3, 2: 2, 3: 1})`。

2. **使用最小堆提取前 k 个高频元素**：
   - `heapq.nlargest(k, count.keys(), key=count.get)` 返回 `[1, 2]`。

### 时间复杂度分析（改进方案）：

- **时间复杂度**：O(n log k)
  - 其中 n 是数组的长度，log k 是维护堆的时间复杂度。

### 空间复杂度分析（改进方案）：

- **空间复杂度**：O(n)
  - 存储频率计数的额外空间。

### 总结：

- **桶排序法**：高效地找到前 k 个高频元素，适合处理频率均匀的情况。
- **最小堆法**：适用于处理频率不均匀的情况，尤其是在 k 较小的情况下。
