### LeetCode 题目：70. 爬楼梯

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        if n == 1: 
            return 1  # 只有一种方法到达第一个台阶
        if n == 2:
            return 2  # 两种方法到达第二个台阶
        
        at_one = 1  # 到达第一个台阶的方式数
        at_two = 2  # 到达第二个台阶的方式数
        total = 0  # 用于存储当前台阶的方式数

        # 从第 3 阶开始计算
        for i in range(3, n + 1):
            total = at_one + at_two  # 当前台阶的方式数等于前一个台阶和前两个台阶的方式数之和
            at_one = at_two  # 更新前一个台阶的方式数
            at_two = total  # 更新前两个台阶的方式数
        
        return total  # 返回总的方式数
```

### 逐行解释：

1. **`def climbStairs(self, n: int) -> int:`**：
   - 定义主函数 `climbStairs`，接受整数 `n`，返回到达第 `n` 阶的不同方法数。

2. **`if n == 1:`**：
   - 检查如果只有一个台阶，返回 1，因为只有一种方法。

3. **`if n == 2:`**：
   - 检查如果有两个台阶，返回 2，因为有两种方法（1+1 或 2）。

4. **`at_one = 1`**：
   - 初始化变量 `at_one`，表示到达第一个台阶的方法数为 1。

5. **`at_two = 2`**：
   - 初始化变量 `at_two`，表示到达第二个台阶的方法数为 2。

6. **`total = 0`**：
   - 初始化变量 `total`，用于存储当前台阶的方法数。

7. **`for i in range(3, n + 1):`**：
   - 从第 3 阶开始循环到第 `n` 阶，计算每个台阶的方法数。

8. **`total = at_one + at_two`**：
   - 当前台阶的方法数等于前一个台阶和前两个台阶的方法数之和。

9. **`at_one = at_two`**：
   - 更新前一个台阶的方式数为当前的前两个台阶的方式数。

10. **`at_two = total`**：
    - 更新前两个台阶的方式数为当前台阶的方式数。

11. **`return total`**：
    - 返回计算出的总方法数。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码需要遍历到第 `n` 阶，因此时间复杂度为 O(n)。

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间来存储变量 `at_one`、`at_two` 和 `total`。

### 总结：

- **动态规划方法**：通过动态规划有效计算到达每个台阶的方法数，利用前两个台阶的方法数来构建当前台阶的方法数。
- **高效性**：该算法在时间和空间上的效率使其非常适合处理大规模输入的情况。！
