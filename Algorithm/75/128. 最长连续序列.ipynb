### 128. 最长连续序列

https://leetcode.com/problems/longest-consecutive-sequence/description/

https://www.youtube.com/watch?v=gbMlwKENRj0&list=PLx2fZyxEuihLSjtEzLKxf20yV8LKLAibL&index=22


#### 问题描述：
给定一个无序的整数数组 `nums`，返回数组中最长连续元素序列的长度。

#### 示例 1：
```python
输入：nums = [100, 4, 200, 1, 3, 2]
输出：4
解释：最长连续序列是 [1, 2, 3, 4]，长度为 4。
```

#### 示例 2：
```python
输入：nums = [0, 0]
输出：1
```

### 原始代码（哈希集合法）：

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)  # 将输入数组转换为集合，方便查找
        longest = 0  # 初始化最长连续序列的长度

        for n in nums:  # 遍历每个数字
            if (n - 1) not in num_set:  # 检查 n 是否是序列的起始点
                nxt = n + 1  # 初始化下一个数字
                while nxt in num_set:  # 查找连续的下一个数字
                    nxt += 1  # 如果找到，继续查找下一个
                longest = max(longest, nxt - n)  # 更新最长连续序列的长度
        
        return longest  # 返回最长连续序列的长度
```

### 逐行解释与示例运行：

1. **`def longestConsecutive(self, nums: List[int]) -> int:`**
   - 定义主函数 `longestConsecutive`，接受一个整数列表 `nums`，返回最长连续序列的长度。

2. **`num_set = set(nums)`**
   - **将输入数组转换为集合**，以便于 O(1) 的查找。

3. **`longest = 0`**
   - **初始化最长连续序列的长度为 0**。

4. **`for n in nums:`**
   - 遍历数组中的每个数字。

5. **`if (n - 1) not in num_set:`**
   - **检查当前数字是否是连续序列的起始点**，即该数字的前一个数字不在集合中。

6. **`nxt = n + 1`**
   - **初始化下一个数字为 `n + 1`**。

7. **`while nxt in num_set:`**
   - **检查当前数字的下一个数字是否在集合中**。

8. **`nxt += 1`**
   - **如果找到下一个数字，序列长度加 1**。

9. **`longest = max(longest, nxt - n)`**
   - **更新最长连续序列的长度**，`nxt - n` 代表当前连续序列的长度。

10. **`return longest`**
    - **返回最长连续序列的长度**。

### 示例运行：

假设我们有以下输入：

```python
nums = [100, 4, 200, 1, 3, 2]
```

运行代码步骤如下：

1. **初始化**：
   - `num_set = {100, 4, 200, 1, 3, 2}`。
   - `longest = 0`。

2. **遍历数字**：
   - `n = 100`：`99` 不在集合中，跳过。
   - `n = 4`：`3` 在集合中，跳过。
   - `n = 200`：`199` 不在集合中，跳过。
   - `n = 1`：`0` 不在集合中，初始化 `nxt = 2`。
     - `2` 在集合中，`nxt = 3`。
     - `3` 在集合中，`nxt = 4`。
     - `4` 在集合中，`nxt = 5`，找不到，结束循环。
     - `longest = max(0, 4 - 1)` → `longest = 4`。
   - `n = 3`：`2` 在集合中，跳过。
   - `n = 2`：`1` 在集合中，跳过。

3. **返回结果**：
   - 返回 `longest = 4`。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 其中 n 是数组的长度。每个数字最多被访问两次：一次在外层循环中，另一次在内层 `while` 循环中。

### 空间复杂度分析：

- **空间复杂度**：O(n)
  - 存储元素的集合需要额外的空间。

### 提示和警告：

1. **检查边界情况**：
   - 对于空数组，应返回 0。

2. **集合使用**：
   - 确保输入数据足够大时，集合的使用是合理的，避免空间浪费。

3. **处理重复元素**：
   - 由于使用了集合，处理重复元素时无需担心。

### 改进方案（使用排序）：

可以使用排序的方法来查找最长连续序列。

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if not nums:  # 检查输入是否为空
            return 0
        
        nums = sorted(set(nums))  # 对数组去重并排序
        longest = 1  # 初始化最长长度为 1
        current_streak = 1  # 初始化当前连续序列长度为 1

        for i in range(1, len(nums)):  # 从第二个元素开始遍历
            if nums[i] == nums[i - 1] + 1:  # 如果当前元素是前一个元素的连续数字
                current_streak += 1  # 当前序列长度加 1
            elif nums[i] != nums[i - 1]:  # 如果当前元素不等于前一个元素且不是连续的
                longest = max(longest, current_streak)  # 更新最长序列长度
                current_streak = 1  # 重置当前序列长度
        
        return max(longest, current_streak)  # 返回最长序列长度
```

### 示例运行（改进方案）：

假设我们有以下输入：

```python
nums = [100, 4, 200, 1, 3, 2]
```

运行代码步骤如下：

1. **初始化**：
   - 检查输入是否为空，非空。
   - `nums = sorted(set(nums))` → `nums = [1, 2, 3, 4, 100, 200]`。

2. **遍历**：
   - `i = 1`，`2 == 1 + 1` → `current_streak = 2`。
   - `i = 2`，`3 == 2 + 1` → `current_streak = 3`。
   - `i = 3`，`4 == 3 + 1` → `current_streak = 4`。
   - `i = 4`，`100 != 4 + 1` → `longest = 4`，`current_streak = 1`。
   - `i = 5`，`200 != 100 + 1` → `longest = 4`，`current_streak = 1`。

3. **返回结果**：
   - 返回 `max(longest, current_streak)` → `4`。

### 时间复杂度分析（改进方案）：

- **时间复杂度**：O(n log n)
  - 排序操作的复杂度是 O(n log n)。

### 空间复杂度分析（改进方案）：

- **空间复杂度**：O(n)
  - 存储去重后的数组需要额外空间。

### 总结：

- **哈希集合法**：高效查找最长连续序列，时间复杂度为 O(n)。
- **排序法**：相对简单，但时间复杂度更高，适合小数据集。
