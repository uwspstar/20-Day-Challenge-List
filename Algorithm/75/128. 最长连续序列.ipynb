### LeetCode 题目：128. 最长连续序列

#### 问题描述：
给定一个未排序的整数数组 `nums`，找出最长的连续序列的长度。

#### 示例 1：
```python
输入：nums = [100, 4, 200, 1, 3, 2]
输出：4
解释：最长的连续序列是 [1, 2, 3, 4]，其长度为 4。
```

#### 示例 2：
```python
输入：nums = [0, 0]
输出：1
```

### 原始代码（使用集合）：

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)  # 将数组元素放入集合以去重并加快查找速度
        length = 0  # 初始化最长连续序列长度
        
        for num in nums:  # 遍历数组中的每个数字
            if num - 1 not in num_set:  # 仅当 num 是序列的起始数字时开始计数
                nxt = num + 1  # 从当前数字开始检查连续序列
                while nxt in num_set:  # 如果下一个数字在集合中，继续增加
                    nxt += 1
                length = max(length, nxt - num)  # 更新最长长度
        
        return length  # 返回最长连续序列长度
```

### 逐行解释：

1. **`def longestConsecutive(self, nums: List[int]) -> int:`**：
   - 定义主函数 `longestConsecutive`，接受一个整数数组 `nums`，返回最长连续序列的长度。

2. **`num_set = set(nums)`**：
   - 将数组转换为集合 `num_set`，以去除重复元素并提高查找效率。

3. **`length = 0`**：
   - 初始化变量 `length`，用于记录最长连续序列的长度。

4. **`for num in nums:`**：
   - 遍历数组中的每个数字。

5. **`if num - 1 not in num_set:`**：
   - 检查 `num` 是否是连续序列的起始数字。如果 `num - 1` 不在集合中，表示 `num` 是一个新序列的开始。

6. **`nxt = num + 1`**：
   - 初始化 `nxt` 为当前数字的下一个数字，用于查找连续序列。

7. **`while nxt in num_set:`**：
   - 当 `nxt` 在集合中时，继续增加 `nxt`，以找到所有连续的数字。

8. **`nxt += 1`**：
   - 将 `nxt` 增加 1，检查下一个数字。

9. **`length = max(length, nxt - num)`**：
   - 更新最长长度，计算当前序列的长度为 `nxt - num`，并与 `length` 比较，保留较大的值。

10. **`return length`**：
    - 返回最长连续序列的长度。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 在最坏情况下，每个数字都被访问两次，因此时间复杂度为 O(n)，其中 n 是数组的长度。

- **空间复杂度**：O(n)
  - 使用集合存储所有元素，空间复杂度为 O(n)。

### 总结：

- **集合方法**：利用集合的查找效率，可以在 O(1) 的时间内检查数字是否存在，进而有效找到最长的连续序列。
- **高效性**：该算法在处理大规模输入时表现良好，能够快速找到结果。
