### LeetCode 题目：3. 无重复字符的最长子串

https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

#### 问题描述：
给定一个字符串 `s`，请你找出其中不含有重复字符的最长子串的长度。

#### 示例 1：
```python
输入：s = "abcabcbb"
输出：3
解释：因为无重复字符的最长子串是 "abc"，所以其长度为 3。
```

#### 示例 2：
```python
输入：s = "bbbbb"
输出：1
解释：因为无重复字符的最长子串是 "b"，所以其长度为 1。
```

### 原始代码：

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        seen = {}
        l = 0
        length = 0
        for r in range(len(s)):
            c = s[r]
            if c in seen and seen[c] >= l:
                l = seen[c] + 1
            else:
                length = max(length, r - l + 1 )
            seen[c] = r
        
        return length
```

### 改进后的代码：

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        seen = {}  # 创建一个字典用于存储字符及其索引
        l = 0  # 左指针初始化
        length = 0  # 记录最长子串长度
        
        for r in range(len(s)):  # 右指针遍历字符串
            c = s[r]  # 当前字符
            if c in seen and seen[c] >= l:  # 如果字符已见且在当前窗口内
                l = seen[c] + 1  # 更新左指针位置
            length = max(length, r - l + 1)  # 更新最长子串长度
            seen[c] = r  # 更新字符的索引
        
        return length  # 返回最长子串的长度
```

### 改进点解释：

   - 代码中去掉了多余的 `else` 分支，使得逻辑更简洁，直接在每次循环中更新长度。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 代码遍历字符串一次，左右指针只向前移动，整体复杂度为 O(n)，其中 n 是字符串 `s` 的长度。

- **空间复杂度**：O(min(n, m))
  - 这里 `n` 是字符串的长度，`m` 是字符集的大小（例如，ASCII 字符集的大小为 128）。在最坏情况下，字典 `seen` 可能存储所有不同的字符。

### 总结：

- **滑动窗口**：通过使用滑动窗口技巧（左指针和右指针），有效地找到不重复字符的最长子串。
- **字典的使用**：利用字典存储字符及其最后出现的位置，使得查找和更新操作的时间复杂度为 O(1)。
