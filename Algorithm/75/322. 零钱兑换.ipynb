### LeetCode 题目：322. 零钱兑换

https://leetcode.com/problems/coin-change/description/

https://www.youtube.com/watch?v=H9bfqozjoqs&t=1s

#### 问题描述：
给定不同面额的硬币 `coins` 和一个总金额 `amount`，计算可以凑成总金额所需的最小硬币数。如果没有任何一种硬币组合能组成总金额，返回 `-1`。

#### 示例 1：
```python
输入：coins = [1, 2, 5], amount = 11
输出：3
解释：11 = 5 + 5 + 1
```

#### 示例 2：
```python
输入：coins = [2], amount = 3
输出：-1
```

### 原始代码（动态规划方法）：

```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        dp = [float('inf')] * (amount + 1)  # 初始化 dp 数组，大小为 amount + 1
        dp[0] = 0  # 零金额需要 0 个硬币

        for coin in coins:  # 遍历每种硬币
            for i in range(coin, amount + 1):  # 更新 dp 数组
                dp[i] = min(dp[i], dp[i - coin] + 1)  # 选择当前硬币
        
        return dp[amount] if dp[amount] != float('inf') else -1  # 返回结果
```

### 逐行解释：

1. **`def coinChange(self, coins: List[int], amount: int) -> int:`**：
   - 定义主函数 `coinChange`，接受一个整数数组 `coins`（硬币面额）和一个整数 `amount`（目标金额），返回所需的最小硬币数。

2. **`dp = [float('inf')] * (amount + 1)`**：
   - 初始化一个动态规划数组 `dp`，大小为 `amount + 1`，初始值为正无穷大（表示初始状态无法达到）。

3. **`dp[0] = 0`**：
   - 对于金额为 0 的情况，所需的硬币数为 0，因此设置 `dp[0] = 0`。

4. **`for coin in coins:`**：
   - 遍历每种硬币面额。

5. **`for i in range(coin, amount + 1):`**：
   - 对于每个硬币，更新从当前硬币面额到目标金额的 `dp` 数组。

6. **`dp[i] = min(dp[i], dp[i - coin] + 1)`**：
   - 更新 `dp[i]`，即当前金额的最小硬币数。通过选择当前硬币，比较当前硬币数与不使用当前硬币的硬币数（`dp[i - coin] + 1`）的最小值。

7. **`return dp[amount] if dp[amount] != float('inf') else -1`**：
   - 如果 `dp[amount]` 仍然是正无穷大，表示无法凑成目标金额，返回 `-1`；否则返回所需的最小硬币数。

### 时间复杂度分析：

- **时间复杂度**：O(n * m)
  - 其中 n 是目标金额 `amount`，m 是硬币种类数。需要遍历每种硬币，并更新每个金额。

- **空间复杂度**：O(n)
  - 动态规划数组 `dp` 的大小为 `amount + 1`，所以空间复杂度为 O(n)。

### 总结：

- **动态规划方法**：通过动态规划有效计算构成目标金额所需的最小硬币数，利用选择当前硬币和不选择的情况来构建状态。
- **高效性**：该算法在时间和空间上的效率使其适合处理较大规模的输入。
