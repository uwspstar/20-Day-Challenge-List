### LeetCode 题目：191. 位1的个数

https://leetcode.com/problems/number-of-1-bits/description/

https://www.youtube.com/watch?v=5Km3utixwZs

#### 问题描述：
编写一个函数，计算二进制中1的个数。

#### 示例 1：
```python
输入：n = 00000000000000000000000000001011
输出：3
解释：输入的二进制表示中有三位是1。
```

#### 示例 2：
```python
输入：n = 00000000000000000000000010000000
输出：1
```

### 改进后的代码：

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        res = 0  # 初始化计数器
        while n:  # 当 n 不为 0 时继续循环
            res += 1  # 每次找到一个1位时，计数增加
            n = n & (n - 1)  # 利用 n & (n - 1) 移除最低位的1
        
        return res  # 返回1的总数
```

### 改进点解释：

1. **移除右移操作**：
   - 右移操作会逐位处理，效率较低。而使用 `n & (n - 1)` 可以直接移除最低位的1，这样效率更高。

2. **清晰的变量命名**：
   - 代码保持了清晰的注释，便于理解每一步的作用。

### 时间复杂度分析：

- **时间复杂度**：O(k)
  - 这里 k 是 n 中1的位数，最坏情况下需要遍历所有1位。

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间来存储 `res` 变量。

### 总结：

- **位运算技巧**：利用位运算中的技巧，能够高效计算二进制中1的个数。
- **高效性**：通过直接移除最低位的1，算法在处理较大数值时仍能保持高效。

### LeetCode 题目：191. 位 1 的个数

https://leetcode.com/problems/number-of-1-bits/description/

#### 问题描述：
编写一个函数，输入为一个无符号整数 `n`，返回其二进制表示中 `1` 的个数。

#### 示例 1：
```python
输入：n = 00000000000000000000000000001011
输出：3
解释：输入的二进制表示中有三位是 '1'。
```

#### 示例 2：
```python
输入：n = 00000000000000000000000010000000
输出：1
解释：输入的二进制表示中有一位是 '1'。
```

### 原始代码（Brian Kernighan 算法）：

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        res = 0  # 初始化结果为 0
        while n > 0:  # 当 n 大于 0 时
            n = n & (n - 1)  # 清除最低位的 1
            res += 1  # 计数增加
        return res  # 返回结果
```

### 逐行解释与示例运行：

1. **`def hammingWeight(self, n: int) -> int:`**
   - 定义主函数 `hammingWeight`，接受一个无符号整数 `n`，返回其二进制表示中 `1` 的个数。

2. **`res = 0`**
   - **初始化结果 `res` 为 0**，用于计数。

3. **`while n > 0:`**
   - 当 `n` 大于 0 时，继续循环。

4. **`n = n & (n - 1)`**
   - **使用位操作清除最低位的 `1`**。这一操作将 `n` 中的最低位 `1` 变为 `0`。例如，`n = 11 (1011)`，`n - 1 = 10 (1010)`，`n & (n - 1) = 2 (0010)`。

5. **`res += 1`**
   - **每次清除一个 `1`，计数加 1**。

6. **`return res`**
   - **返回 `res`，即 `1` 的个数**。

### 示例运行：

假设我们有以下输入：

```python
n = 11  # 二进制表示为 1011
```

运行代码步骤如下：

1. **初始化**：
   - `res = 0`，`n = 11`。

2. **循环执行**：
   - 第一次循环：
     - `n = 11`，`11 & (11 - 1) = 10`，更新 `n = 10`，`res = 1`。
   - 第二次循环：
     - `n = 10`，`10 & (10 - 1) = 8`，更新 `n = 8`，`res = 2`。
   - 第三次循环：
     - `n = 8`，`8 & (8 - 1) = 0`，更新 `n = 0`，`res = 3`。

3. **返回结果**：
   - 循环结束时，返回 `res = 3`。

### 时间复杂度分析：

- **时间复杂度**：O(k)
  - 其中 k 是 `n` 的二进制表示中 `1` 的个数。循环次数取决于 `1` 的个数，而不是 `n` 的大小。

### 空间复杂度分析：

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间。

### 提示和警告：

1. **无符号整数**：
   - 确保输入为无符号整数，以避免负数对结果的影响。

2. **二进制表示**：
   - 理解二进制的运算，特别是位运算的使用，以确保算法的正确性。

3. **边界情况**：
   - 处理输入为 `0` 的情况，返回 `0`。

### 改进方案（逐位检查）：

可以使用逐位检查的方法来计算 `1` 的个数，虽然效率稍低，但逻辑更直观。

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        res = 0  # 初始化结果为 0
        for i in range(32):  # 遍历 32 位
            res += (n >> i) & 1  # 检查每一位是否为 1
        return res  # 返回结果
```

### 示例运行（改进方案）：

假设我们有以下输入：

```python
n = 11  # 二进制表示为 1011
```

运行代码步骤如下：

1. **初始化**：
   - `res = 0`。

2. **逐位检查**：
   - 遍历 32 位：
     - 对于 `i = 0`：`n >> 0` 是 `11`，最后一位是 `1`，`res = 1`。
     - 对于 `i = 1`：`n >> 1` 是 `5`，最后一位是 `1`，`res = 2`。
     - 对于 `i = 2`：`n >> 2` 是 `2`，最后一位是 `0`，`res = 2`。
     - 对于 `i = 3`：`n >> 3` 是 `1`，最后一位是 `1`，`res = 3`。
     - 其他位均为 `0`，保持 `res = 3`。

3. **返回结果**：
   - 返回 `res = 3`。

### 时间复杂度分析（改进方案）：

- **时间复杂度**：O(1)
  - 固定的 32 次迭代，不受 `n` 的值影响。

### 空间复杂度分析（改进方案）：

- **空间复杂度**：O(1)
  - 只使用了常量级的额外空间。

### 总结：

- **位操作法**：高效计算 `1` 的个数，时间复杂度为 O(k)。
- **逐位检查法**：更直观，但时间复杂度为 O(1)。
