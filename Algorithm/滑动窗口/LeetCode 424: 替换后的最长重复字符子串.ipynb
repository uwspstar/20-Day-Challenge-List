### LeetCode 题目：424. 替换后的最长重复字符

https://leetcode.com/problems/longest-repeating-character-replacement/description/
https://www.youtube.com/watch?v=gqXU1UyA8pk

#### 问题描述：
给定一个只包含大写字母的字符串 `s`，你可以将最多 `k` 个字符替换为同一个字符。找到可以获得的最长重复字符的长度。

#### 示例 1：
```python
输入：s = "ABAB", k = 2
输出：4
解释：用两个 'A' 替换两个 'B'，得到 "AAAA"。
```

#### 示例 2：
```python
输入：s = "AABABBA", k = 1
输出：4
解释：用一个 'B' 替换为 'A'，得到 "AABBBBA"。
```

### 原始代码（滑动窗口法）：

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}  # 字符频率计数
        l = 0  # 左指针
        ans = 0  # 最大长度初始化为 0

        for r in range(len(s)):  # 右指针遍历字符串
            count[s[r]] = count.get(s[r], 0) + 1  # 更新字符频率

            # 检查当前窗口的大小与最大频率的关系
            while (r - l + 1) - max(count.values()) > k:
                count[s[l]] -= 1  # 移除左指针字符的频率
                l += 1  # 移动左指针
            
            ans = max(ans, r - l + 1)  # 更新最大长度
        
        return ans  # 返回结果
```

### 逐行解释与示例运行：

1. **`def characterReplacement(self, s: str, k: int) -> int:`**
   - 定义主函数 `characterReplacement`，接受一个字符串 `s` 和一个整数 `k`，返回最长替换后的字符长度。

2. **`count = {}`**
   - **初始化一个字典 `count`**，用于记录每个字符的频率。

3. **`l = 0`**
   - **初始化左指针 `l`，指向当前窗口的开始位置**。

4. **`ans = 0`**
   - **初始化最大长度 `ans` 为 0**。

5. **`for r in range(len(s)):`**
   - 遍历字符串，右指针 `r` 从 0 到字符串长度。

6. **`count[s[r]] = count.get(s[r], 0) + 1`**
   - **更新当前字符 `s[r]` 的频率**，使用 `get` 方法确保当字符不存在时返回 0。

7. **`while (r - l + 1) - max(count.values()) > k:`**
   - **检查当前窗口的大小减去最大字符频率是否大于 k**。如果是，则需要缩小窗口。

8. **`count[s[l]] -= 1`**
   - **减少左指针指向字符的频率**。

9. **`l += 1`**
   - **移动左指针，缩小窗口**。

10. **`ans = max(ans, r - l + 1)`**
    - **更新最大长度**，即当前窗口的大小。

11. **`return ans`**
    - **返回最大长度**。

### 示例运行：

假设我们有以下输入：

```python
s = "AABABBA"
k = 1
```

运行代码步骤如下：

1. **初始化**：
   - `count = {}`, `l = 0`, `ans = 0`。

2. **遍历字符串**：
   - `r = 0`，`s[r] = 'A'`：
     - `count = {'A': 1}`，窗口大小 `1 - 0 + 1 = 1`，最大频率 `1`，符合条件，更新 `ans = 1`。
   - `r = 1`，`s[r] = 'A'`：
     - `count = {'A': 2}`，窗口大小 `2 - 0 + 1 = 2`，最大频率 `2`，符合条件，更新 `ans = 2`。
   - `r = 2`，`s[r] = 'B'`：
     - `count = {'A': 2, 'B': 1}`，窗口大小 `3 - 0 + 1 = 3`，最大频率 `2`，符合条件，更新 `ans = 3`。
   - `r = 3`，`s[r] = 'A'`：
     - `count = {'A': 3, 'B': 1}`，窗口大小 `4 - 0 + 1 = 4`，最大频率 `3`，符合条件，更新 `ans = 4`。
   - `r = 4`，`s[r] = 'B'`：
     - `count = {'A': 3, 'B': 2}`，窗口大小 `5 - 0 + 1 = 5`，最大频率 `3`，不符合条件，进入 `while` 循环：
       - `count['A'] -= 1`，`l += 1`，更新 `count = {'A': 2, 'B': 2}`。
   - `r = 5`，`s[r] = 'B'`：
     - `count = {'A': 2, 'B': 3}`，窗口大小 `5 - 1 + 1 = 5`，最大频率 `3`，不符合条件，进入 `while` 循环：
       - `count['B'] -= 1`，`l += 1`，更新 `count = {'A': 2, 'B': 2}`。
   - `r = 6`，`s[r] = 'A'`：
     - `count = {'A': 3, 'B': 2}`，窗口大小 `5 - 2 + 1 = 5`，最大频率 `3`，不符合条件，进入 `while` 循环：
       - `count['A'] -= 1`，`l += 1`，更新 `count = {'A': 1, 'B': 2}`。

3. **返回结果**：
   - 最后返回 `ans = 4`。

### 时间复杂度分析：

- **时间复杂度**：O(n)
  - 其中 n 是字符串的长度。每个字符最多被访问两次（一次由右指针，一次由左指针）。

### 空间复杂度分析：

- **空间复杂度**：O(1)
  - 使用一个固定大小的字典来记录字符频率，不受输入大小影响。
